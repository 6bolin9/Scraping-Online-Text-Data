```python
import requests
from bs4 import BeautifulSoup
```

```python
##### collect the URL of the first web page with Requests
#####assign the URL for the first page to the variable page by using the method requests.get().
page = requests.get('https://www.nga.gov/Collection/artists.html?const_startLetter=z&pageNumber=1&lastFacet=const_startLetter')
```

```python
##### Create a BeautifulSoup object
#####parses it from Python’s built-in html.parser
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup)
```

```python
# Pull all text from the BodyText div
artist_name_list = soup.find(class_='navreturns-btm')
print(artist_name_list)
```

```python
#####Pull text from all instances of <a> tag within  div
artist_name_list_items = artist_name_list.find_all(class_='art')
print(artist_name_list_items)
```

```python
#Next, at the bottom of our program file, we will want to create a for loop in order to iterate over all the artist names that we just put into the artist_name_list_items variable.

#We’ll print these names out with the prettify() method in order to turn the Beautiful Soup parse tree into a nicely formatted Unicode string.

```

```python
for artist_name in artist_name_list_items:
    print(artist_name.prettify())
```

```python
##access only the actual artists’ names, target the contents of the <a> tags rather than print out the entire link tag.
##We can do this with Beautiful Soup’s .contents----- return the tag’s children as a Python list data type.
for artist_name in artist_name_list_items:
    names = artist_name.contents[0]
    print(names)
```

```python
###Capture the URLs associated with those artists? 
####extract URLs found within a page’s <a> tags by using Beautiful Soup’s get('href') method.
for artist_name in artist_name_list_items:
    names = artist_name.contents[0]
    links = 'https://www.nga.gov' + artist_name.get('href')
    print(names)
    print(links)
```

```python
import requests
import csv
from bs4 import BeautifulSoup


f = csv.writer(open('z-artist-names.csv', 'w'))
f.writerow(['Name', 'Link'])

###注意整个代码有两个for loop
###1）产生每个URL，整合成组
###2）抓取每个URL，成组存储
pages = []  #网页(url)【】空白

for i in range(1, 6):
    url = 'https://www.nga.gov/collection/anZ' + str(i) + '.htm'
    pages.append(url)##网页组(pages)产生，（1）for loop产生

for item in pages:###针对网页组(pages)的每个URL(item)在进行抓取，（2）for loop产生
    page = requests.get(item)###request.get(item)
    soup = BeautifulSoup(page.text, 'html.parser')###beautifulsoup(requests.get(item).text, 'html.parser')

    last_links = soup.find(class_='AlphaNav')######
    last_links.decompose()

    artist_name_list = soup.find(class_='BodyText')######
    artist_name_list_items = artist_name_list.find_all('a')

    for artist_name in artist_name_list_items:#####forloop 存储
        names = artist_name.contents[0]   #####.content[0]，存储第一个text内容
        links = 'https://www.nga.gov' + artist_name.get('href')#####.get('href')，存储其中的网址

        f.writerow([names, links])####都保存到file:f
        

#the first for loop is iterating over the pages
#second for loop is scraping data from each of those pages
#then is adding the artists’ names and links line by line through each row of each page
```
